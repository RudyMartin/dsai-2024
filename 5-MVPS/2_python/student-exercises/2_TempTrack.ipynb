To calculate the standard deviation of the temperatures in the given data, we'll go through each step of the calculation. The standard deviation measures the amount of variation or dispersion in a set of values. Here are the steps to manually calculate the standard deviation:

### Steps to Calculate Standard Deviation:
1. **Calculate the Mean (Average):**
   - The mean is the average of the numbers in a dataset. You sum up all the numbers and then divide by the count of numbers.

2. **Calculate Each Deviation from the Mean:**
   - For each number in the dataset, subtract the mean and square the result (the squared difference).

3. **Find the Mean of These Squared Differences:**
   - Add up all of the squared differences and then divide by the number of values in the dataset.

4. **Take the Square Root:**
   - The square root of the step above gives you the standard deviation.

### Calculation:
Let's calculate the standard deviation for the temperatures given in the data [22, 24, 19, 24, 27, 30, 23].

#### Step 1: Calculate the Mean
Mean =
$$
\frac{22 + 24 + 19 + 24 + 27 + 30 + 23}{7} = \frac{169}{7} \approx 24.14
$$

#### Step 2: Calculate Each Deviation from the Mean and Square It
- $(22 - 24.14)^2 \approx 4.58$
- $(24 - 24.14)^2 \approx 0.02$
- $(19 - 24.14)^2 \approx 26.46$
- $(24 - 24.14)^2 \approx 0.02$
- $(27 - 24.14)^2 \approx 8.14$
- $(30 - 24.14)^2 \approx 34.32$
- $(23 - 24.14)^2 \approx 1.30$

#### Step 3: Find the Mean of These Squared Differences
Mean of Squared Differences =
$$
\frac{4.58 + 0.02 + 26.46 + 0.02 + 8.14 + 34.32 + 1.30}{7} \approx 10.55
$$

#### Step 4: Take the Square Root
Standard Deviation =
$$
\sqrt{10.55} \approx 3.25
$$
