## Simple set of python scripts to chart and help explain Algorithmic Complexity using BIG O notation.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the input size
n = np.linspace(1, 100, 100)

# Define the different complexity classes
O_1 = np.ones_like(n)
O_log_n = np.log(n)
O_n = n
O_n_log_n = n * np.log(n)
O_n2 = n ** 2
O_2n = 2 ** n
O_n_factorial = np.array([np.math.factorial(int(i)) for i in n])

# Plot the complexity classes
plt.figure(figsize=(10, 6))

plt.plot(n, O_1, label="O(1) : Constant", color='cyan')
plt.plot(n, O_log_n, label="O(logN):  Logarithmic", color='purple')
plt.plot(n, O_n, label="O(N) : Linear", color='red')
plt.plot(n, O_n_log_n, label="O(NlogN) : Lineararithmetic", color='brown')
plt.plot(n, O_n2, label="O(N²) : Quadratic", color='green')
plt.plot(n, O_2n, label="O(2^N) : Exponential", color='yellow')
plt.plot(n, O_n_factorial, label="O(N!) :  Factorial", color='blue')

# Add labels and legend
plt.xlabel("input")
plt.ylabel("steps")
plt.title("Big-O Complexity Classes")
plt.ylim(0, 100)  # Limit the y-axis for better visualization
plt.legend()

# Show plot
plt.show()
```
